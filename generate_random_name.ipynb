```python

from faker import Faker

# Initialize the Faker instance
fake = Faker()

# Define a function to generate random names
def generate_random_names(num_names=5):
    random_names = [fake.name() for _ in range(num_names)]
    return random_names

# Call the function to generate and print 5 random names
if __name__ == "__main__":
    num_names = int(input("How many random names would you like to generate? "))
    names = generate_random_names(num_names)
    for name in names:
        print(name)
```
